```python
#cek apakah ada data outlier dalam data harapan hidup
sns.boxplot(new_data.Harapan_hidup)
new_data[new_data.Harapan_hidup<50]
```

```python
#deteksi data outlier pada kolom kesehatan
q1_kesehatan=new_data.Kesehatan.quantile(0.25)
q3_kesehatan=new_data.Kesehatan.quantile(0.75)
iqr_kesehatan=q3_kesehatan-q1_kesehatan
batas_atas_kesehatan=q3_kesehatan+(1.5*iqr_kesehatan)
batas_bawah_kesehatan=q1_kesehatan-(1.5*iqr_kesehatan)
new_data[(new_data.Kesehatan>=batas_atas_kesehatan)|(new_data.Kesehatan<=batas_bawah_kesehatan)]
```

```python
#deteksi data outlier pada kolom harapan Harapan_hidup
q1_harapan=new_data.Harapan_hidup.quantile(0.25)
q3_harapan=new_data.Harapan_hidup.quantile(0.75)
iqr_harapan=q3_harapan-q1_harapan
batas_atas_harapan=q3_harapan+(1.5*iqr_harapan)
batas_bawah_harapan=q1_harapan-(1.5*iqr_harapan)
new_data[(new_data.Harapan_hidup>=batas_atas_harapan)|(new_data.Harapan_hidup<=batas_bawah_harapan)]
```

```python
#handling data outlier kesehatan
median_kesehatan=new_data.Kesehatan.median()
persentil_kesehatan=new_data.Kesehatan.quantile(0.95)
new_data.Kesehatan=np.where(new_data.Kesehatan>persentil_kesehatan,median_kesehatan,new_data.Kesehatan)
```

```python
new_data[(new_data.Kesehatan>=batas_atas_kesehatan)|(new_data.Kesehatan<=batas_bawah_kesehatan)]
```

```python
#handling outlier data harapan hidup
new_data['harapan hidup']=np.where(new_data.Harapan_hidup>batas_atas_harapan,batas_atas_harapan,np.where(new_data.Harapan_hidup<batas_bawah_harapan,batas_bawah_harapan,new_data.Harapan_hidup))
sns.boxplot(new_data['harapan hidup'])
```

```python
#lakukan scaling pada data negara help
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

scale=StandardScaler()
data_scale=scale.fit_transform(new_data[['Kesehatan','harapan hidup']])
data_scale
```

```python
#menentukan kluster pada dataset dengan metode silhoutte score
silhoutte_scores=[]
jumlah_cluster=range(2,12)
for i in jumlah_cluster:
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(new_data[['Kesehatan','harapan hidup']])
  labels=kmeans.labels_
  silhoutte_scores.append(silhouette_score(new_data[['Kesehatan','harapan hidup']],labels,metric='euclidean'))
```

```python
#visualisasi grafik silhouette
plt.figure(figsize=(10,15))
sns.lineplot(x=jumlah_cluster,y=silhoutte_scores)
sns.scatterplot(x=jumlah_cluster,y=silhoutte_scores)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new_data['Kesehatan'],new_data['harapan hidup'])
```

```python
#kmeans dengan menggunakan optimal cluster pada silhouette score
kmean=KMeans(n_clusters=2)
kmean.fit(new_data[['Kesehatan','harapan hidup']])
new_data['cluster']=kmean.labels_
new_data
```

```python
#visualisasi hasil cluster menggunakan metode silhoutte
plt.figure(figsize=(10,5))
sns.scatterplot(x=new_data.Kesehatan,y=new_data['harapan hidup'],hue=new_data['cluster'])
plt.show()
```

```python
#berdasarkan visualisasi tampilkan 3 negara yang memiliki angka harapan hidup dibawah 50
new_data[(new_data['harapan hidup']<50) & (new_data.Kesehatan<12) & (new_data['cluster']==1)]
```

```python
#prediksi data dengan angka kesehatan 8.5 dan harapan hidup 50
kmean.predict(new_data[['Kesehatan','harapan hidup']])
```
